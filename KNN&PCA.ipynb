```python
'''
Shobhit-Sinha
Theoretical

1. What is K-Nearest Neighbors (KNN) and how does it work?

What is K-Nearest Neighbors (KNN)?

K-Nearest Neighbors (KNN) is a supervised machine learning algorithm used for both classification and regression tasks. It's a simple, yet powerful 
algorithm that makes predictions based on the similarity of a data point to its neighbors in the feature space.

How does KNN work?

Here's a breakdown of how the KNN algorithm works:

1. Choosing 'k': The algorithm starts by selecting a value for 'k', which represents the number of nearest neighbors to consider. 'k' is a 
   hyperparameter that needs to be tuned for optimal performance.
2. Calculating distances: For a new data point that needs to be classified or predicted, the algorithm calculates the distance between this point and
   all other data points in the training dataset. Various distance metrics can be used, such as Euclidean distance, Manhattan distance, or Minkowski 
   distance.
3. Identifying nearest neighbors: Based on the calculated distances, the algorithm identifies the 'k' data points that are closest to the new data
   point. These are the 'k' nearest neighbors.
4. Making predictions (Classification): In classification tasks, the algorithm assigns the new data point to the class that is most frequent among its
  'k' nearest neighbors. For example, if k=5 and 3 out of the 5 nearest neighbors belong to class 'A', then the new data point will be classified
   as 'A'.
5. Making predictions (Regression): In regression tasks, the algorithm predicts the value of the target variable for the new data point by averaging the
   values of the target variable for its 'k' nearest neighbors.

Example:

Imagine you have a dataset of fruits with features like size, color, and weight. You want to classify a new fruit based on its features using KNN.

1. You choose a value for 'k', say k=3.
2. The algorithm calculates the distance between the new fruit and all other fruits in the dataset.
3. It identifies the 3 nearest neighbors to the new fruit.
4. If 2 out of the 3 nearest neighbors are apples and 1 is an orange, the algorithm would classify the new fruit as an apple.

Key considerations:

. Choosing the right value for 'k' is crucial. A small 'k' can make the model sensitive to noise, while a large 'k' can smooth out local patterns and
  lead to overfitting.
. Data preprocessing is important. Features should be scaled to prevent features with larger values from dominating the distance calculations.
. KNN can be computationally expensive for large datasets as it needs to calculate distances between all data points.
'''
```

```python
'''
3. What is the role of the distance metric in KNN?
In KNN, the distance metric plays a crucial role in determining the similarity between data points. It is used to calculate the distance between the
new data point and all other data points in the training dataset. This distance is then used to identify the 'k' nearest neighbors to the new data 
point.

The choice of distance metric can significantly impact the performance of the KNN algorithm. Different distance metrics are suitable for different types
of data and applications.

Here are some commonly used distance metrics in KNN:

. Euclidean Distance: This is the most commonly used distance metric. It calculates the straight-line distance between two points in a multi-dimensional
  space. It is suitable for continuous numerical data.
. Manhattan Distance: This metric calculates the distance between two points by summing the absolute differences of their coordinates. It is also known
  as the L1 distance or taxicab distance. It is suitable for discrete or categorical data.
. Minkowski Distance: This is a generalized distance metric that includes both Euclidean and Manhattan distance as special cases. It is controlled by a
  parameter 'p'. When p=2, it becomes Euclidean distance, and when p=1, it becomes Manhattan distance.
. Hamming Distance: This metric is used for categorical data and calculates the number of positions at which two strings differ. It is suitable for text
  or sequence data.
. Cosine Similarity: This metric measures the angle between two vectors. It is often used in text analysis and information retrieval.

The choice of distance metric depends on the nature of the data and the specific problem. For example, if the data is continuous and numerical, 
Euclidean distance might be a good choice. If the data is categorical, Manhattan or Hamming distance might be more appropriate.

Here's how the distance metric affects the KNN algorithm:

1. Identifying nearest neighbors: The distance metric is used to calculate the distance between the new data point and all other data points in the
   training dataset. The 'k' data points with the smallest distances are considered the nearest neighbors.
2. Making predictions: The predictions in KNN are based on the values of the target variable for the nearest neighbors. The distance metric influences
   which data points are considered as nearest neighbors, and therefore, it affects the final prediction.
'''
```

```python
'''
4. What is the Curse of Dimensionality in KNN?

The Curse of Dimensionality refers to the phenomenon where the performance of the KNN algorithm (and other distance-based algorithms) degrades as the 
number of features or dimensions in the dataset increases. This degradation occurs because the data points become increasingly sparse in 
high-dimensional spaces, making it difficult to find meaningful nearest neighbors.

Here's a breakdown of how the Curse of Dimensionality affects KNN:
1. Increased Sparsity: As the number of dimensions increases, the data points become more spread out in the feature space. This sparsity makes it harder
   to find data points that are close to each other, leading to a decrease in the density of data points.
2. Distance Metric Ineffectiveness: In high-dimensional spaces, the distance between data points tends to become less meaningful. All data points start
   to appear equidistant from each other, making it difficult for the distance metric to effectively distinguish between neighbors and non-neighbors.
3. Computational Cost: The computational cost of calculating distances between data points increases exponentially with the number of dimensions. This
  makes KNN computationally expensive for high-dimensional datasets.
4. Overfitting: In high-dimensional spaces, the risk of overfitting increases. The model may become too complex and start to memorize the training data,
  leading to poor generalization performance on unseen data.

How to mitigate the Curse of Dimensionality in KNN:

1. Dimensionality Reduction: Techniques like Principal Component Analysis (PCA) or Linear Discriminant Analysis (LDA) can be used to reduce the number
   of features while preserving the most important information.
2. Feature Selection: Carefully selecting the most relevant features for the task can help to reduce the dimensionality of the dataset.
3.Using appropriate distance metrics: Some distance metrics are more robust to high-dimensional data than others. For example, cosine similarity can be 
  a better choice than Euclidean distance in high-dimensional spaces.
4. Increasing the value of 'k': Using a larger value for 'k' can help to smooth out the effects of noise and sparsity in high-dimensional data.
'''
```

```python
'''
5. How can we choose the best value of K in KNN?

Choosing the optimal value for 'k' is crucial for the performance of the KNN algorithm. A well-chosen 'k' can balance the trade-off between overfitting
and underfitting, leading to a model that generalizes well to unseen data. Here are some techniques for selecting the best value of K:

1. Cross-Validation:

. k-Fold Cross-Validation: Divide the dataset into 'k' folds. Train the KNN model on k-1 folds and validate it on the remaining fold. Repeat this
  process 'k' times, using a different fold for validation each time. Calculate the average performance metric (e.g., accuracy, F1-score) across all
  folds for each value of 'k'.
. Leave-One-Out Cross-Validation (LOOCV): A special case of k-fold cross-validation where k is equal to the number of data points. Train the model on 
  all data points except one and validate it on the left-out data point. Repeat this for all data points and calculate the average performance metric
  for each value of 'k'.

2. Grid Search:

. Define a range of possible values for 'k' (e.g., 1 to 20).
. Use cross-validation to evaluate the performance of the KNN model for each value of 'k' within the defined range.
. Select the value of 'k' that yields the best performance metric on the validation set.

3. Heuristics:

. Square root of the number of data points: A common rule of thumb is to set 'k' to the square root of the number of data points in the training set.
. Odd values for 'k' in classification: When using KNN for classification, it's often recommended to choose odd values for 'k' to avoid ties in the
  majority voting process.
4. Visualization:

. Error Rate vs. k Plot: Plot the error rate (or other performance metric) against different values of 'k'. Look for the value of 'k' where the error
  rate is minimized or the performance metric is maximized.
. Elbow Method: In the error rate vs. k plot, look for an "elbow" point where the error rate starts to decrease at a slower rate. This point can indicate a good value for 'k'.

Considerations:

. Computational Cost: Cross-validation can be computationally expensive, especially for large datasets. Consider using techniques like grid search with
  a smaller range of 'k' values to reduce the computational burden.
. Bias-Variance Trade-off: A smaller 'k' can lead to a more flexible model with lower bias but higher variance. A larger 'k' can lead to a less flexible
  model with higher bias but lower variance. The optimal 'k' balances this trade-off.
. Problem-Specific Considerations: The best value of 'k' can vary depending on the specific dataset and problem. Experimentation and careful evaluation
  are crucial for finding the optimal value.
'''
```

```python
'''
6. What are KD Tree and Ball Tree in KNN?

KD Tree and Ball Tree are data structures used to accelerate the nearest neighbor search in KNN. They help to reduce the computational cost of searching
for the 'k' nearest neighbors in large datasets.

KD Tree (k-dimensional tree):

. A binary tree that recursively partitions the data space along different dimensions.
. Each node in the tree represents a hyperplane that splits the data into two subspaces.
. The tree is built by iteratively selecting a dimension and splitting the data along that dimension based on a chosen threshold value.
. During nearest neighbor search, the tree is traversed to quickly narrow down the search space and identify potential nearest neighbors.

Ball Tree:

. A hierarchical data structure that partitions the data space into hyperspheres (balls).
. Each node in the tree represents a ball that encloses a subset of data points.
. The tree is built by recursively splitting the data into two balls based on a chosen centroid and radius.
. During nearest neighbor search, the tree is traversed to quickly prune away balls that are unlikely to contain nearest neighbors.

Comparison:

Feature	                      KD Tree	                               Ball Tree
Partitioning	             Hyperplanes	                          Hyperspheres
Construction	           Axis-aligned splits	                     Centroid-based splits
Search Efficiency	     Efficient for low-dimensional data	      Efficient for high-dimensional data
Space Complexity	           Lower	                                   Higher

When to use which:

. KD Tree: Preferred for low-dimensional data (typically less than 20 dimensions) where the data is relatively uniformly distributed.
. Ball Tree: Preferred for high-dimensional data where the data may be clustered or non-uniformly distributed.

Benefits of using KD Tree and Ball Tree in KNN:

. Reduced Search Time: They significantly reduce the number of distance calculations needed to find the nearest neighbors, leading to faster search
  times.
. Improved Efficiency: They make KNN more efficient for large datasets, enabling the algorithm to scale to larger datasets.
'''
```

```python
'''
8. What are the disadvantages of KNN?
K-Nearest Neighbors (KNN) is a simple and intuitive machine learning algorithm, but it does have several disadvantages:

1. Computationally Expensive:

. KNN requires calculating the distance between a query point and all the points in the training dataset. This becomes particularly slow as the size of
  the dataset grows, since it is essentially a brute-force approach to finding neighbors.

2. High Memory Usage:

. Since KNN is a lazy learner, it does not build a model during the training phase but stores the entire training dataset. This can result in high 
memory usage, especially for large datasets.

3. Curse of Dimensionality:

. In high-dimensional spaces (many features), the concept of "closeness" between points becomes less meaningful. As the number of dimensions increases, 
  the distance between points grows, and it becomes harder for KNN to distinguish meaningful neighbors, leading to poor performance.

4. Sensitive to Irrelevant Features:

. KNN is sensitive to the inclusion of irrelevant or redundant features. If the dataset contains many features that don't contribute to the class 
  prediction, they can negatively affect the algorithm's performance.

5. Sensitive to Scaling:

. KNN uses distance metrics (like Euclidean distance) to find neighbors. If the features in the dataset have different scales (e.g., one feature ranges
  from 1 to 1000, and another from 0 to 1), the algorithm might give more importance to the features with larger numerical ranges unless the data is 
  properly normalized or standardized.

6. Class Imbalance:

. KNN can struggle with imbalanced datasets. If one class is significantly more frequent than others, the algorithm may be biased towards predicting the
  majority class because it relies on counting neighbors.

7. Noisy Data:

. KNN can be heavily influenced by noise and outliers in the dataset. Since the classification of a point depends on its neighbors, noisy or outlier 
  points can distort the classification result.

8. Lack of Interpretability:

. Unlike some models (like decision trees), KNN doesn't offer easy interpretability in terms of how decisions are made. While it's easy to understand
  that a query point is classified based on its nearest neighbors, understanding why specific neighbors were chosen is harder, especially in 
  high-dimensional data.
'''
```

```python
'''
10. What is PCA (Principal Component Analysis)?

Principal Component Analysis (PCA) is a statistical technique used for dimensionality reduction while preserving as much variability (information) in
the data as possible. It transforms the data into a new coordinate system such that the first few new dimensions (called principal components) capture
the most significant variation in the data. PCA is commonly used for feature extraction, data visualization, and improving the efficiency of machine
learning algorithms by reducing the number of features (or dimensions) in a dataset.

Here’s a breakdown of how PCA works:
Key Concepts:
1. Principal Components:

. Principal components are new features (linear combinations of the original features) that capture the maximum variance in the data.

. The first principal component (PC1) captures the largest possible variance, the second principal component (PC2) captures the second-largest variance,
  and so on.

. Each principal component is orthogonal (uncorrelated) to the others.

2. Variance Maximization:

. PCA aims to find directions (principal components) that maximize the variance in the data. This means that the new dimensions should explain the most
  information possible about the dataset.

3. Linear Transformation:

. PCA performs a linear transformation to convert the original features into a set of new features (principal components). This transformation is done 
  using eigenvectors and eigenvalues of the covariance matrix of the data.

Steps Involved in PCA:
1. Standardize the Data :

. It’s crucial to standardize the data (subtract the mean and divide by the standard deviation) if the features have different units or scales, because 
  PCA is sensitive to the variance of each feature.

2. Compute the Covariance Matrix:

. The covariance matrix measures how the features in the dataset vary with respect to each other. The diagonal elements represent the variances of each
  feature, while the off-diagonal elements represent the covariances between pairs of features.

3. Calculate Eigenvalues and Eigenvectors:

. Eigenvectors of the covariance matrix represent the directions (principal components), while eigenvalues indicate the magnitude of the variance
  captured by each principal component.

4. Sort Eigenvectors by Eigenvalues:

. Sort the eigenvectors in decreasing order of their corresponding eigenvalues. The eigenvector with the largest eigenvalue is the first principal 
  component, the one with the second-largest eigenvalue is the second principal component, and so on.

5. Select the Top k Principal Components:

. Choose the first k eigenvectors (principal components) that capture the most significant variance in the data. This step reduces the dimensionality of
  the dataset by selecting the most important components while discarding less informative ones.

6. Transform the Data:

. Finally, the data is projected onto the selected principal components. This transforms the original dataset into a new space with fewer dimensions
 (if dimensionality reduction is desired).

Benefits of PCA:
. Dimensionality Reduction: PCA helps reduce the number of features in a dataset, which can improve computational efficiency and reduce overfitting in 
  machine learning models.

. Noise Reduction: By removing less important components, PCA can reduce the noise in the data, leading to potentially better generalization.

. Data Visualization: PCA can be used to reduce high-dimensional data to 2 or 3 dimensions for visualization purposes, helping to identify patterns or
  clusters visually.

. Improved Performance: For some algorithms, reducing the number of features while preserving variance can improve performance by eliminating redundant 
  or irrelevant features.

When to Use PCA:
. High-dimensional data: If you have many features (high-dimensional space) and want to reduce it to a lower-dimensional representation without losing
  important information.

. Visualizing data: If you want to plot high-dimensional data in 2D or 3D to understand patterns or structure in the data.

. Noise reduction: When you want to remove noise or irrelevant features that may not contribute significantly to the data’s variance.

Limitations of PCA:

. Linearity: PCA assumes linear relationships between features. It may not perform well if the relationships in the data are nonlinear.

. Interpretability: The new principal components might not always be easy to interpret, especially if they are combinations of multiple original 
  features.

. Sensitivity to scaling: PCA is sensitive to the scale of the features, so data should be standardized before applying PCA.
'''
```

```python
'''
11. How does PCA work?

1. Standardization:

. PCA begins by standardizing the dataset. This involves subtracting the mean of each feature and dividing by its standard deviation.
. Standardization ensures that features with different scales don't disproportionately influence the analysis.
. Reasoning: Without standardization, features with larger values would dominate the analysis, even if they are not necessarily more important.

2. Covariance Matrix Calculation:

. After standardization, the covariance matrix is calculated.
. This matrix shows the relationships between different features in the dataset.
. The covariance matrix helps identify patterns and dependencies among the features.

3. Eigenvalue and Eigenvector Calculation:

. Eigenvalues and eigenvectors are derived from the covariance matrix.
. Eigenvectors represent the principal components, which are directions of maximum variance in the data.
. Eigenvalues indicate the amount of variance explained by each principal component.
. Reasoning: Eigenvectors and eigenvalues provide a way to capture the most important information in the dataset and reduce its dimensionality.

4. Selection of Principal Components:

. Principal components are ranked based on their corresponding eigenvalues.
. The components with the highest eigenvalues capture the most variance and are selected.
. The number of principal components to retain depends on the desired level of dimensionality reduction.
. Reasoning: By selecting the most important principal components, we can significantly reduce the dimensionality of the data while preserving most of
  the important information.

5. Data Transformation:

. The original dataset is transformed by projecting it onto the selected principal components.
. This creates a new dataset with fewer dimensions while retaining the most important information from the original dataset.
. Reasoning: The transformed dataset is easier to analyze and visualize, making it suitable for various machine learning tasks.
'''
```

```python
'''
15. How do you decide the number of components to keep in PCA?
Methods for Choosing the Number of Components

Scree Plot:

Reasoning: This plot shows the explained variance by each principal component. You look for the "elbow" point, where the variance explained starts to
drop off significantly. This point suggests the optimal number of components to retain.

Code Example:
'''
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris

# Load data
iris = load_iris()
X = iris.data

# Fit PCA
pca = PCA()
pca.fit(X)

# Create scree plot
plt.plot(np.arange(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_)
plt.xlabel("Number of Components")
plt.ylabel("Explained Variance Ratio")
plt.title("Scree Plot")
plt.show()
```

```python
'''
To see the output, run the code. Looking for the elbow point in the plot will help in deciding the number of components.

2. Cumulative Explained Variance:

. Reasoning: This method involves selecting enough components to explain a desired percentage of the total variance (e.g., 95%).

Code Example:
'''
import numpy as np
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris

# Load data
iris = load_iris()
X = iris.data

# Fit PCA
pca = PCA()
pca.fit(X)

# Calculate cumulative explained variance
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)

# Find number of components explaining 95% variance
n_components = np.argmax(cumulative_variance >= 0.95) + 1

print("Number of components explaining 95% variance:", n_components)
```

```python
'''
16. Can PCA be used for classification?
Principal Component Analysis (PCA) is primarily a dimensionality reduction technique, not a classification algorithm. It works by transforming the 
original features into a new set of orthogonal features called principal components, which capture the maximum variance in the data.

However, PCA can be used as a preprocessing step for classification tasks. Here's how:

How PCA Helps in Classification:
1. Dimensionality Reduction: High-dimensional data can lead to overfitting and increased computational costs. PCA reduces the number of features while
   retaining the most important information.

2. Noise Reduction: By focusing on the components with the highest variance, PCA helps eliminate noise, which can improve classifier performance.

3. Visualization: PCA can project data into 2D or 3D for visualization, helping understand data distribution before applying classification algorithms.

PCA + Classifier Workflow:

. Apply PCA to the dataset to reduce dimensions.

. Train a classifier (e.g., SVM, Logistic Regression, KNN) on the PCA-transformed data.

. Evaluate the classifier's performance.

Example:
. Before PCA: You have 100 features, and a classifier struggles with overfitting.

. After PCA: You reduce to 10 principal components, and the classifier performs better with faster training.
'''
```

```python
'''
17. What are the limitations of PCA?
While Principal Component Analysis (PCA) is a powerful tool, it has several limitations, especially when applied to real-world data. Here are the key ones:

1. Loss of Interpretability

. Issue: PCA transforms original features into principal components, which are linear combinations of the original features. This makes it hard to
  interpret what each component represents in terms of the original variables.

. Impact: In domains like healthcare or finance, where interpretability is crucial, this can be a problem.

2. Sensitivity to Scaling
. Issue: PCA is sensitive to the scale of the data. Features with larger ranges dominate the principal components.

. Solution: Always standardize the data (mean = 0, variance = 1) before applying PCA, especially when features have different units (e.g., age vs.
  income).

3. Assumes Linearity
. Issue: PCA assumes that the data relationships are linear, which limits its effectiveness with complex, non-linear patterns.

. Workaround: For non-linear data, techniques like Kernel PCA or t-SNE can be more effective.

4. Unsupervised Nature (Ignores Labels)
. Issue: PCA doesn’t consider class labels. It focuses solely on variance, which may not align with class separability.

. Impact: In classification tasks, PCA might reduce dimensions in a way that hurts the model’s ability to distinguish between classes.

. Alternative: Linear Discriminant Analysis (LDA) considers class information when reducing dimensions.

 5. Computational Complexity for Large Datasets
. Issue: PCA involves eigenvalue decomposition, which can be computationally intensive for very large datasets with thousands of features.

. Solution: For large datasets, use Incremental PCA or Randomized PCA, which are faster approximations.

6. Risk of Overfitting (With Too Many Components)
. Issue: If too many principal components are retained, the model might still overfit, especially in noisy datasets.

. Best Practice: Use techniques like the explained variance ratio to choose the optimal number of components.

 7. Not Robust to Outliers
. Issue: PCA is sensitive to outliers since they can heavily influence the direction of the principal components.

. Fix: Apply robust PCA methods or preprocess the data to handle outliers before applying PCA.

'''
```

```python
'''
18. How do KNN and PCA complement each other?
KNN and PCA: A Synergistic Combination

K-Nearest Neighbors (KNN) and Principal Component Analysis (PCA) can work together very effectively in machine learning tasks. Here's how they complement
each other:

1. PCA Reduces Dimensionality for KNN:

. KNN's Weakness: KNN's performance can be negatively affected by high-dimensional data (the "curse of dimensionality"). Distances between points become
  less meaningful as the number of dimensions increases, making it difficult for KNN to find the true nearest neighbors.
. PCA's Solution: PCA reduces the dimensionality of the data by identifying the most important features (principal components). This helps to mitigate
  the curse of dimensionality and improve KNN's efficiency and accuracy.

2. PCA Can Improve KNN's Speed:

. KNN's Computational Cost: KNN can be computationally expensive, especially with large datasets, because it needs to calculate distances between the
  query point and all data points in the training set.

. PCA's Speed Boost: By reducing the number of features, PCA reduces the computational cost of distance calculations, making KNN faster.

3. PCA Can Remove Noise and Redundancy:

. Noisy Data: Real-world datasets often contain noise and redundant features that can hinder KNN's performance.
. PCA's Cleaning Effect: PCA can help to filter out noise and irrelevant information, allowing KNN to focus on the most important patterns in the data.
'''
```

```python
'''
20. What are the key differences between PCA and Linear Discriminant Analysis (LDA)?
PCA vs. LDA: A Comparison

Both Principal Component Analysis (PCA) and Linear Discriminant Analysis (LDA) are dimensionality reduction techniques, but they have different goals
and assumptions:

Feature	                          PCA	                                          LDA
Type	                      Unsupervised	                                   Supervised
Goal	                  Maximize variance in the data	                 Maximize class separability
Focus	             Finding directions of maximum variance	        Finding directions that best discriminate between classes
Input	                    Features only	                              Features and class labels
Output	                  Principal components	                           Discriminant components
When to Use	     Data exploration, noise reduction, feature extraction	Classification tasks, feature extraction for classification

Key Differences Explained:

1. Supervision:

. PCA is unsupervised, meaning it doesn't use class labels. It treats all data points equally and aims to find the directions of maximum variance in the
  data.
. LDA is supervised and requires class labels. It uses this information to find the directions that best separate different classes.

2. Goal:

. PCA aims to maximize the variance explained by the principal components. It tries to capture as much information as possible in fewer dimensions.
. LDA aims to maximize the separation between different classes. It tries to find the directions that project the data into a lower-dimensional space
  where the classes are well-separated.

3. Focus:

. PCA focuses on finding the directions of maximum variance in the data, regardless of class labels.
. LDA focuses on finding the directions that best discriminate between classes, even if those directions don't capture the most variance overall.

4. Output:

. PCA produces principal components, which are linear combinations of the original features.
. LDA produces discriminant components, which are also linear combinations of the original features, but are chosen to maximize class separability.

5. When to Use:

. PCA is often used for data exploration, noise reduction, and feature extraction when you don't have class labels or when you want to reduce 
  dimensionality without considering class information.
. LDA is typically used for classification tasks or when you want to extract features that are specifically relevant for classifying data into different
  categories.

'''
```

```python
'''
24. Train a KNN Classifier with different values of K and visualize decision boundaries
'''
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler

# Step 1: Load the Iris dataset and select only the first two features for easy visualization
data = load_iris()
X = data.data[:, :2]  # First two features: sepal length and sepal width
y = data.target

# Step 2: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 3: Standardize the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Step 4: Define a function to plot the decision boundary
def plot_decision_boundary(X, y, model, ax, title="KNN Decision Boundary"):
    h = .02  # Step size in the meshgrid
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

    # Create a meshgrid for plotting
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plot the contour and data points
    ax.contourf(xx, yy, Z, alpha=0.8)
    scatter = ax.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', s=50, cmap=plt.cm.Paired)
    ax.set_title(title)
    return scatter

# Step 5: Train and plot decision boundaries for different values of K
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# Values of K to test
k_values = [1, 3, 5, 10]

# Iterate through each K value
for i, k in enumerate(k_values):
    # Create a KNN model for the current K
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train, y_train)
    
    # Plot the decision boundary for this K
    ax = axes[i // 2, i % 2]
    plot_decision_boundary(X_test, y_test, model, ax, title=f"KNN Decision Boundary (K={k})")

# Add a color legend for the plot
fig.subplots_adjust(hspace=0.4, wspace=0.4)
plt.show()
```

```python
'''
32. Train a KNN Regressor and analyze the effect of different K values on performance.
'''
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score

# 1. Load the California housing dataset (for regression)
california = fetch_california_housing()
X = california.data
y = california.target

# 2. Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. Initialize a list to store performance metrics for different K values
k_values = range(1, 21)  # We'll try K values from 1 to 20
mse_scores = []
r2_scores = []

# 4. Loop over different K values and train KNN Regressor
for k in k_values:
    knn_regressor = KNeighborsRegressor(n_neighbors=k)
    knn_regressor.fit(X_train, y_train)
    
    # Predict on test set
    y_pred = knn_regressor.predict(X_test)
    
    # Calculate Mean Squared Error (MSE) and R² Score
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    mse_scores.append(mse)
    r2_scores.append(r2)

# 5. Plot the results
plt.figure(figsize=(14, 6))

# Plot MSE
plt.subplot(1, 2, 1)
plt.plot(k_values, mse_scores, marker='o', linestyle='-', color='b')
plt.title('Effect of K on Mean Squared Error (MSE)')
plt.xlabel('K (Number of Neighbors)')
plt.ylabel('Mean Squared Error (MSE)')

# Plot R² Score
plt.subplot(1, 2, 2)
plt.plot(k_values, r2_scores, marker='o', linestyle='-', color='r')
plt.title('Effect of K on R² Score')
plt.xlabel('K (Number of Neighbors)')
plt.ylabel('R² Score')

plt.tight_layout()
plt.show()

# Display best K value based on MSE
best_k = k_values[np.argmin(mse_scores)]
print(f"Best K based on MSE: {best_k}")

```

```python
'''
38. Train a PCA model and analyze the effect of different numbers of components on accuracy.
'''
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# 1. Load the Iris dataset (for classification)
iris = load_iris()
X = iris.data
y = iris.target

# 2. Split the dataset into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. Initialize a KNN classifier
knn = KNeighborsClassifier(n_neighbors=3)

# 4. Train KNN classifier with different numbers of PCA components
components_range = range(1, X.shape[1] + 1)  # Try from 1 to 4 components for Iris dataset
accuracies = []

for n_components in components_range:
    # Apply PCA to reduce dimensions
    pca = PCA(n_components=n_components)
    X_train_pca = pca.fit_transform(X_train)
    X_test_pca = pca.transform(X_test)
    
    # Train the KNN classifier
    knn.fit(X_train_pca, y_train)
    
    # Predict and calculate accuracy
    y_pred = knn.predict(X_test_pca)
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)

# 5. Plot the effect of different numbers of components on accuracy
plt.figure(figsize=(8, 6))
plt.plot(components_range, accuracies, marker='o', linestyle='-', color='b')
plt.title("Effect of Number of PCA Components on KNN Accuracy")
plt.xlabel("Number of Principal Components")
plt.ylabel("Accuracy")
plt.xticks(components_range)
plt.grid(True)
plt.show()

# Print accuracies for different numbers of components
for n_components, accuracy in zip(components_range, accuracies):
    print(f"Number of components: {n_components}, Accuracy: {accuracy:.4f}")
```

```python
'''
43. Train a KNN Classifier and evaluate using ROC-AUC score.
'''
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
from sklearn.metrics import RocCurveDisplay

# Step 1: Load the Iris Dataset
iris = load_iris()
X = iris.data
y = iris.target

# Step 2: Binarize the target labels for ROC-AUC calculation (multiclass problem)
# For ROC-AUC, we need to convert labels to binary format for each class
y_bin = label_binarize(y, classes=[0, 1, 2])

# Step 3: Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y_bin, test_size=0.3, random_state=42)

# Step 4: Train the KNN Classifier
model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train, y_train)

# Step 5: Predict probabilities for each class
y_pred_prob = model.predict_proba(X_test)

# Step 6: Compute ROC curve and AUC score for each class
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(3):  # There are 3 classes in the Iris dataset
    fpr[i], tpr[i], _ = roc_curve(y_test[:, i], y_pred_prob[i][:, 1])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Step 7: Plot ROC curves
plt.figure(figsize=(10, 8))

for i in range(3):
    plt.plot(fpr[i], tpr[i], color='blue', lw=2, label=f'Class {i} (AUC = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], color='gray', linestyle='--')  # Random classifier line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for KNN Classifier')
plt.legend(loc='lower right')
plt.show()

# Output the AUC score for each class
for i in range(3):
    print(f"Class {i} - AUC: {roc_auc[i]:.2f}")
```

```python
'''
45. Train a KNN Classifier and perform feature selection before training.
'''
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.feature_selection import RFE
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Step 1: Load the Iris Dataset
iris = load_iris()
X = iris.data
y = iris.target

# Step 2: Perform Feature Selection using Recursive Feature Elimination (RFE)
# Initialize a Decision Tree classifier (which has feature importance)
dt = DecisionTreeClassifier(random_state=42)

# Perform RFE for feature selection (let's select the top 2 features)
rfe = RFE(estimator=dt, n_features_to_select=2)
X_selected = rfe.fit_transform(X, y)

# Print selected features
print(f"Selected Features (Indices): {rfe.support_}")
print(f"Selected Features (Names): {iris.feature_names}")

# Step 3: Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.3, random_state=42)

# Step 4: Train the KNN Classifier on selected features
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

# Step 5: Make Predictions and Evaluate the Model
y_pred = knn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of KNN Classifier after Feature Selection: {accuracy:.2f}")

# Step 6: Visualize the Selected Features' Impact (Optional)
# We will visualize the data using the selected features (only 2 features left after RFE)
plt.figure(figsize=(8, 6))
plt.scatter(X_selected[:, 0], X_selected[:, 1], c=y, cmap=plt.cm.RdYlBu, edgecolor='k', s=150, alpha=0.7)
plt.title('KNN Classifier with Selected Features (RFE)')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.colorbar()
plt.show()
```

```python
'''
46. Train a PCA model and visualize the data reconstruction error after reducing dimensions.
'''
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.metrics import mean_squared_error

# Step 1: Load the Iris Dataset
iris = load_iris()
X = iris.data

# Step 2: Standardize the data (important for PCA)
# PCA is sensitive to the scale of the data, so we need to standardize it first
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Apply PCA to reduce dimensions (e.g., to 2 components)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Step 4: Reconstruct the original data from the reduced PCA components
X_reconstructed = pca.inverse_transform(X_pca)

# Step 5: Calculate the Reconstruction Error (Mean Squared Error)
reconstruction_error = mean_squared_error(X_scaled, X_reconstructed)
print(f"Reconstruction Error (MSE): {reconstruction_error:.4f}")

# Step 6: Visualize the reconstruction error for each data point
# Calculate the per-sample reconstruction error
errors = np.linalg.norm(X_scaled - X_reconstructed, axis=1)

# Create a scatter plot of the reconstruction errors
plt.figure(figsize=(8, 6))
plt.scatter(range(len(errors)), errors, color='blue', alpha=0.7)
plt.title("Reconstruction Error for Each Sample")
plt.xlabel("Sample Index")
plt.ylabel("Reconstruction Error (Euclidean Norm)")
plt.show()
```

```python
'''
47. Train a KNN Classifier and visualize the decision boundary.
'''
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# Step 1: Generate a 2D synthetic dataset for visualization
X, y = make_classification(n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, random_state=42)

# Step 2: Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 3: Train the KNN Classifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# Step 4: Visualize the decision boundary
# Create a meshgrid over the feature space
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))

# Predict the class labels for the entire grid
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundary
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.RdYlBu)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.RdYlBu, edgecolors='k', marker='o', s=100, label='Data points')

# Add labels and title
plt.title('KNN Classifier Decision Boundary')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend(loc='upper right')
plt.show()
```

```python
'''
48. Train a PCA model and analyze the effect of different numbers of components on data variance.
'''
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

# Step 1: Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Step 2: Standardize the data (important for PCA)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Apply PCA with different numbers of components
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Step 4: Plot the explained variance for each component
explained_variance_ratio = pca.explained_variance_ratio_

# Plot the explained variance for each component
plt.figure(figsize=(8, 6))
plt.bar(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, alpha=0.7, color='blue')
plt.title('Explained Variance by Each Principal Component')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.xticks(range(1, len(explained_variance_ratio) + 1))
plt.show()

# Step 5: Plot the cumulative explained variance
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)

plt.figure(figsize=(8, 6))
plt.plot(range(1, len(cumulative_variance_ratio) + 1), cumulative_variance_ratio, marker='o', color='green')
plt.title('Cumulative Explained Variance')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.xticks(range(1, len(cumulative_variance_ratio) + 1))
plt.grid(True)
plt.show()

# Print the explained variance and cumulative variance
print(f"Explained Variance by each component: {explained_variance_ratio}")
print(f"Cumulative Explained Variance: {cumulative_variance_ratio}")
```
